IMPORTING THE REQUIRED LIBRARIES

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import confusion_matrix, precision_score, recall_score
```

IMPORING THE DATASET

```python
df = pd.read_csv("Heart_Disease_Prediction.csv")
df.head()
```

PERFORMING THE EDA

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
### checking the missing values
df.isnull().sum()
```

```python
df.drop("Slope of ST" , inplace = True , axis = 1)
df.drop("EKG results" , inplace = True , axis = 1)
df.drop("ST depression" , inplace = True , axis = 1)
df.drop("Exercise angina" , inplace = True , axis = 1)
```

```python
df["Thallium"] = df["Thallium"].replace({3 : 1 , 6 : 2 , 7 : 3})
df["Heart Disease"] = df["Heart Disease"].replace({"Presence" : 1 , "Absence" : 0})
```

AGE DISTRIBUTION

```python
plt.figure(figsize = (7,5))
sns.histplot(df["Age"], bins = 25 , kde = True)
plt.title("AGE DISTRIBUTION")
plt.show()
```

```python
plt.figure(figsize = (7,5))
sns.scatterplot(x = df["Age"] , y = df["Heart Disease"] , hue=df["Heart Disease"] , color = "viridis")
plt.xlabel("AGE")
plt.ylabel("Heart Disease")
plt.title("AGE VS DISEASE PEOPLE")
plt.show()
```

SCALING THE REQUIRED COLUMNS VALUES

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()
df["Age"] = scaler.fit_transform(df[["Age"]])
df["BP"] = scaler.fit_transform(df[["BP"]])
df["Cholesterol"] = scaler.fit_transform(df[["Cholesterol"]])
df["Max HR"] = scaler.fit_transform(df[["Max HR"]])
df.head()
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(df.corr(numeric_only = True) , cmap = "coolwarm" , annot = True)
plt.title("RELATION BETWEEN EACH FEATURE")
plt.xticks(rotation = 90)
plt.show()
```

CLASSIFICATION OF MALE AND FEMALE ATTACKED BY DISEASE

```python
plt.figure(figsize = (7,5))
sns.countplot(x = df["Sex"] ,hue = df["Heart Disease"]  ,  data = df)
plt.title("AGE DISTRIBUTION")
plt.show()
```

SPLITING THE COLUMNS AND PERFORMING TRAINING , TESTING

```python
x = df[["Age" , "Sex" , "Chest pain type" ,"BP" ,"Cholesterol" , "FBS over 120", "Max HR" , "Number of vessels fluro" , "Thallium"]]
y = df["Heart Disease"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
print("Training samples :",x_train.shape[0])
print("Testing samples :", x_test.shape[0])
```

```python
model = LogisticRegression() 
model.fit(x_train , y_train)
predict = model.predict(x_test)
value = pd.DataFrame({"Actual" : y_test , "Predict" : predict})
value.head()
```

ENSURING THE CONFUSION MATRIX  , PROBABILITY , PRECISION , RECALL , ACCURACY

```python
probability = model.predict_proba(x_test)
cmatrix = confusion_matrix(y_test,predict)
accuracy = ((cmatrix[0][0] + cmatrix[1][1])/(np.sum(cmatrix)))*100
pscore = precision_score(y_test,predict)
rscore = recall_score(y_test,predict)
f1_score = 2*((pscore * rscore)/(pscore + rscore))
print("confusion_matrix :\n",cmatrix)
print("Accuracy :",accuracy)
print("precision_score :",pscore)
print("recall_score :",rscore)
print("f1_score :",f1_score)
```

```python
plt.figure(figsize = (7,4))
sns.heatmap(cmatrix , annot = True , cmap = "coolwarm")
plt.title("RELATION OF THE CONFUSION MATRIX")
plt.show()
```

TESTING THE MODLE WITH NEW DATA

```python
new_data = np.array([[0.856732, 0, 3, 0.214589, -0.987321, 1, 0.643210, 2, 0]])
predict = model.predict(new_data)
print("DISEASE" if predict[0] == 1 else "NOT DISEASE")
```
